```python
import pandas as pd
import matplotlib.pyplot as plt

X_Train = 4610  # 전체 학습 데이터 수

# CSV 파일 경로
files_001bar = {
    "1bar": "./O2_0.001bar흡착량이용/Scripts/Active_Learning_O2_1bar_Score.csv",
    "5bar": "./O2_0.001bar흡착량이용/Scripts/Active_Learning_O2_5bar_Score2.csv",
    "15bar": "./O2_0.001bar흡착량이용/Scripts/Active_Learning_O2_15bar_Score2.csv",
}

files_05bar = {
    "1bar": "./O2_0.5bar흡착량이용/Scripts/Active_Learning_O2_1bar_Score.csv",
    "5bar": "./O2_0.5bar흡착량이용/Scripts/Active_Learning_O2_5bar_Score.csv",
    "15bar": "./O2_0.5bar흡착량이용/Scripts/Active_Learning_O2_15bar_Score.csv",
}
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'   # Windows 기준 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False   
# 샘플링 비율 계산
sampling_rates = [((10 * i + int(X_Train * 0.05)) * 100 / X_Train) for i in range(100)]

# 압력별 Plot
for pressure in ["1bar", "5bar", "15bar"]:
    df_001 = pd.read_csv(files_001bar[pressure])
    df_05 = pd.read_csv(files_05bar[pressure])

    # # --- MSE Plot ---
    # plt.figure(figsize=(8, 5))
    # plt.plot(sampling_rates, df_001["Active_Learning_MSE"], label="0.001bar 기반", color='blue')
    # plt.plot(sampling_rates, df_05["Active_Learning_MSE"], label="0.5bar 기반", color='red')
    # plt.xlabel("샘플링 비율 (%)")
    # plt.ylabel("MSE")
    # plt.title(f"{pressure} - Active Learning MSE")
    # plt.grid(True)
    # plt.legend()
    # plt.tight_layout()
    # plt.show()

    # --- R² Plot ---
    plt.figure(figsize=(8, 5))
    plt.plot(sampling_rates, df_001["Active_Learning_R2"], label="0.001bar 기반", color='blue')
    plt.plot(sampling_rates, df_05["Active_Learning_R2"], label="0.5bar 기반", color='red')
    plt.ylim(0, 1)
    plt.xlim(0, 100)
    plt.xlabel("샘플링 비율 (%)")
    plt.ylabel("R²")
    plt.title(f"{pressure} - Active Learning R²")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
