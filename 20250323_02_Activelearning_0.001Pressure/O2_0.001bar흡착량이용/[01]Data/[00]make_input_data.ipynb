```python
def replace_0_001(COREMOF_O2_X):
    new_df = pd.merge(COREMOF_O2_X , COREMOF_O2_0_001, how = "inner" , on = "Adsorbent")
    new_df["0.5bar_GCMC"] = new_df["Uptake (mol/kg)"]
    new_df.columns = ['Unnamed: 0_x', 'Adsorbent', 'LCD', 'PLD', 'LFPD', 'cm3_g',
        'ASA_m2_cm3', 'ASA_m2_g', 'NASA_m2_cm3', 'NASA_m2_g', 'AV_VF',
        'AV_cm3_g', 'NAV_cm3_g', 'Has_OMS', '0.001bar_GCMC', 'Adsorbate',
        'Uptake', 'Temperature', 'Molecular weight (g/mol)', 'H-bond donors',
        'H-bond acceptors', 'Dipole moment (D)', 'Molecular surface area (Å^2)',
        'Critical temperature (K)', 'Compressibility factor (상온,상압)',
        'Hmolar (상온,상압)', 'Compressibility factor 주어진 압력, 온도',
        'Hmola 주어진 압력, 온도', 'Boiling point at 1atm (K)',
        'Melting point at 1atm (K)', 'Boiling point at 주어진압력 (K)',
        'Melting point at 주어진압력 (K)', 'Pressure', 'Unnamed: 0_y', 'Index',
        'Directory', 'Uptake (mol/kg)']
    new_df["0.001bar_GCMC"] = new_df["0.001bar_GCMC"].str.replace("..", ".").astype(float)
    new_df["0.001bar_GCMC"]
    new_df = new_df[['Adsorbent', 'LCD', 'PLD', 'LFPD', 'cm3_g',
        'ASA_m2_cm3', 'ASA_m2_g', 'NASA_m2_cm3', 'NASA_m2_g', 'AV_VF',
        'AV_cm3_g', 'NAV_cm3_g', 'Has_OMS', '0.001bar_GCMC', 'Adsorbate',
        'Uptake', 'Temperature', 'Molecular weight (g/mol)', 'H-bond donors',
        'H-bond acceptors', 'Dipole moment (D)', 'Molecular surface area (Å^2)',
        'Critical temperature (K)', 'Compressibility factor (상온,상압)',
        'Hmolar (상온,상압)', 'Compressibility factor 주어진 압력, 온도',
        'Hmola 주어진 압력, 온도', 'Boiling point at 1atm (K)',
        'Melting point at 1atm (K)', 'Boiling point at 주어진압력 (K)',
        'Melting point at 주어진압력 (K)', 'Pressure']]
    return new_df
COREMOF_O2_1_minus1e3_based = replace_0_001(COREMOF_O2_1)
COREMOF_O2_5_minus1e3_based = replace_0_001(COREMOF_O2_5)
COREMOF_O2_15_minus1e3_based = replace_0_001(COREMOF_O2_15)
COREMOF_O2_1_minus1e3_based.to_csv("COREMOF_O2_1_minus1e3_based.csv")
COREMOF_O2_5_minus1e3_based.to_csv("COREMOF_O2_5_minus1e3_based.csv")
COREMOF_O2_15_minus1e3_based.to_csv("COREMOF_O2_15_minus1e3_based.csv")
```
